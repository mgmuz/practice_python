````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# サンプルデータを作成
np.random.seed(0)
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': np.random.randint(1, 100, 5),
    'Scores': np.random.random(5) * 100
}
df = pd.DataFrame(data)
print(df)
````

<img src="images/ex1_table.png" width="301"> <img src="images/ex1_table_detail.png" width="610">


Indexは自動で割り振られる行番号，Category,Values,Scoresは列名(column)である．

作成したデータdfを用いて色々なグラフを描画してみる．

````python
# カテゴリごとの値を棒グラフで表示
plt.figure(figsize=(5, 3)) # figsizeで画像の大きさを決定する
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.title('Bar Plot of Values by Category')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
````

````python
# 箱ひげ図（ボックスプロット）を作成
plt.figure(figsize=(5, 3))
sns.boxplot(x=df['Category'], y=df['Scores'])
plt.title('Box Plot of Scores by Category')
plt.show()
````

````python
# 散布図を作成
plt.figure(figsize=(5, 3))
plt.scatter(df['Values'], df['Scores'], color='green', s=100) # sは丸サイズ
plt.title('Scatter Plot of Values vs. Scores')
plt.xlabel('Values')
plt.ylabel('Scores')
plt.show()
````

````python
# ヒストグラムを作成
plt.figure(figsize=(5, 3))
sns.histplot(df['Scores'], kde=True, color='purple') # kdeは線で描画されるカーネル密度分布
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
````

# Numpyデータを用いてグラフを描画する 
sin波$A・sin(2πft + \phi)$とcos波$A・cos(2πft + \phi)$をグラフ描画する．

````python
import numpy as np
import matplotlib.pyplot as plt

# 周波数、振幅、位相を変数として設定
A = 1       # 振幅
f = 1       # 周波数 (Hz)
phi = 0     # 位相（ラジアン）
sampling_rate = 1000  # サンプリング周波数 (Hz)

# 時間の範囲を設定 (0～2秒の間で生成)
t = np.linspace(0, 2, sampling_rate)

sin_y = A * np.sin(2 * np.pi * f * t + phi) # A * sin(2πft + φ)
cos_y = A * np.cos(2 * np.pi * f * t + phi) # A * cos(2πft + φ)

# 最初の10データのみ出力して確認する
print(sin_y[:10])
print(cos_y[:10])
````

````python
# グラフを表示
plt.figure(figsize=(5, 3))
plt.plot(t, sin_y, c="red", label="sine")
plt.plot(t, cos_y, c="blue", label="cosine")

# タイトル、軸ラベルを設定
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
````
